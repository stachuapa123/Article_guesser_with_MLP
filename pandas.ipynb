```python
# row
df = pd.DataFrame({'Word': ['apple', 'car'], 'Category': ['fruit', 'vehicle']})
new_record_list = ['dog', 'animal']

# 1. Turn the list into a 1-row DataFrame
new_row_df = pd.DataFrame([new_record_list], columns=df.columns)

# 2. Concatenate (glue) them together
df = pd.concat([df, new_row_df], ignore_index=True)

print(df)
```
